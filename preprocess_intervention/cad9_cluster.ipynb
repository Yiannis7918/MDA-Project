```python
import geopandas as gpd
import pandas as pd
```

```python
cad9=pd.read_parquet("/Users/hy/StaDS/23-24 spring/mda/project/MDA/MDA/preprocess_intervention/cad9_clean.parquet",engine="pyarrow")
```

```python
cad9.head(10)
```

```python
cad9['service geometry']=gpd.points_from_xy(cad9['longitude permanence'],cad9['latitude permanence'])
cad9['intervention geometry']=gpd.points_from_xy(cad9['longitude intervention'],cad9['latitude intervention'])
```

```python
cad_cluster=cad9[['province','intervention geometry','service geometry','cityname intervention','citysectionname intervention','province invervention']]
```

```python
cad_cluster_melted = pd.melt(cad_cluster, id_vars=['province','cityname intervention', 'citysectionname intervention','province invervention'],
                         value_vars=['service geometry', 'intervention geometry'],
                    var_name='geometry type', value_name='geometry')
cad_cluster_melted
```

```python
cad_cluster_melted=gpd.GeoDataFrame(cad_cluster_melted, geometry='geometry')
cad_cluster_melted.set_crs(epsg=4326, inplace=True)
```

```python
folder_path="/your path"
cad_cluster_melted.to_file(folder_path+"cad_cluster.shp")
```
