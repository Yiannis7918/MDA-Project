```python
intvn_bxl1['t0']=pd.to_datetime(intvn_bxl1['t0'],utc=True)
intvn_bxl1['t6']=pd.to_datetime(intvn_bxl1['t6'])
```

```python
periods={
    (8, 10): 'morning',
    (11, 13): 'noon',
    (14, 16): 'afternoon',
    (17, 20): 'evening',
    (21, 23): 'night',
    (0, 7): 'midnight'
}
def map_to_period(time):
    for period_range, period_name in periods.items():
        if period_range[0] <= time.hour <= period_range[1]:
            return period_name
    return None
intvn_bxl1['time_period']=intvn_bxl1['t0'].apply(map_to_period)
```
