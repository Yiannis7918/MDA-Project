```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

## import external shape file

```python
# import shapefile

be_shape = gpd.read_file("/Users/camillecu/Downloads/KUL/MDA/project/clustering/vector_shape/be.shp")
```

```python
be_shape
```

```python
intervention = pd.read_parquet("/Users/camillecu/Downloads/KUL/MDA/project/preprocess_intervention/intvn_clean.parquet")

# create a subset only containing the columns we need: 'latitude intervention', 'longitude intervention', 'eventtype firstcall', 'eventlevel trip','same event type'
intervention_subset = intervention[['latitude intervention', 'longitude intervention', 'eventtype firstcall', 'eventtype trip'
                                    ,'same event type']]

# create geometry column from cardiac
intervention_subset = gpd.GeoDataFrame(intervention_subset, 
                                       geometry=gpd.points_from_xy(intervention_subset['longitude intervention'], 
                                                                   intervention_subset['latitude intervention']))
```

```python
intervention_subset
```

```python
# intervention_subset = intervention_subset.set_geometry('geometry')
intervention_subset.set_crs(epsg=4326, inplace=True)
```

```python
# print unique values of 'eventtype firstcall'
#intervention_subset['eventtype firstcall'].unique()
```

```python
#  we only need rows with 'P039 - Cardiac problem (other than thoracic pain)', 'P003 - Cardiac arrest'
cardiac = intervention_subset[intervention_subset['eventtype firstcall'].isin(['P039 - Cardiac problem (other than thoracic pain)', 'P003 - Cardiac arrest'])]
cardiac
```

```python
# Perform a spatial join to find ambulances within each municipality
cardiac_bel = gpd.sjoin(cardiac, be_shape, how='left', op='within')

# Group by municipality name and count the number of ambulances in each municipality
cardiac_counts = cardiac_bel.groupby('mun_name_n').size().reset_index(name='cardiac_count')

# Print the counts
print(cardiac_counts)

```

```python
be_shape = be_shape.merge(cardiac_counts, left_on='mun_name_n', right_on='mun_name_n', how='left')
be_shape = be_shape.fillna(0)
```

```python
# Perform a spatial join to find ambulances within each municipality
intervention_bel = gpd.sjoin(intervention_subset, be_shape, how='left', op='within')

# Group by municipality name and count the number of ambulances in each municipality
intervention_counts = intervention_bel.groupby('mun_name_n').size().reset_index(name='intervention_counts')

be_shape = be_shape.merge(intervention_counts, left_on='mun_name_n', right_on='mun_name_n', how='left')

be_shape = be_shape.fillna(0)
```

```python
be_shape
```

```python
# create a new column 'intervention_rate' which is the ratio of 'intervention_counts' to 'vectors'
be_shape['intervention_rate'] = be_shape['intervention_counts'] / (be_shape['ambulance_'] + be_shape['mug_count']+ be_shape['pit_count'] + 1)
be_shape['cardiac_rate'] = be_shape['cardiac_count'] / (be_shape['vectors'] + 1)
```

```python
import folium
# Initialize a base map with Folium, centered on Belgium
base_map = folium.Map(location=[50.5, 4], zoom_start=7)

# generate a choropleth map with the number of AEDs in each municipality
folium.Choropleth(
    geo_data=be_shape,
    name='choropleth',
    data=be_shape,
    columns=['mun_name_n', 'cardiac_rate'],
    key_on='feature.properties.mun_name_n',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Cardiac rate in each municipality'
).add_to(base_map)

# add tooltip to the map
style_function = lambda x: {'fillColor': 'lightgreen', 'color': 'black', 'fillOpacity': 0.1, 'weight': 0.1}
highlight_function = lambda x: {'fillColor': 'green', 'color': 'green', 'fillOpacity': 0.50, 'weight': 0.1}
folium.GeoJson(
    be_shape,
    style_function=style_function,
    control=False,
    highlight_function=highlight_function,
    tooltip=folium.features.GeoJsonTooltip(
        fields=['mun_name_n', 'cardiac_rate'],
        aliases=['Municipality', 'Cardiac Rate'],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;"),
        sticky=True
    )
).add_to(base_map)
# Add a layer control to toggle layers
folium.LayerControl().add_to(base_map)

# Display the map
#base_map

```
